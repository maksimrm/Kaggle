```python
import datawig
import pandas as pd
import numpy as np
```

```python
data_train = pd.read_pickle("../titanic/data_train.pk")
data_test = pd.read_pickle("../titanic/data_test.pk")
```

```python
data_train.drop(columns = ["Survived", "PassengerId"], inplace=True)
data_test.drop(columns = ["PassengerId"], inplace=True)
```

```python
data_train.info()
```

```python
data_test.info()
```

```python
data = data_train.append(data_test, ignore_index=True)
```

```python
data["Fare"] = data["Fare"].fillna(data["Fare"].mean())
```

```python
data_first = data.copy()
```

```python
data.drop(columns=["Embarked", "third_name"], inplace=True)
```

```python
data["corpus"][data["corpus"] == "N"] = np.nan
data["number_of_cabin"][data["number_of_cabin"] == 0] = np.nan
```

```python
data.info()
```

```python
impute_col = ["Age", "corpus", "number_of_cabin"]
```

```python
info_columns = ['Pclass', 'Sex',  'SibSp', 'Parch', 'Fare',  
        'first_name', 'second_name', 'title', 'ticket']
```

```python
df_train, df_test = datawig.utils.random_split(data)
```

```python
imputer = datawig.SimpleImputer(
    input_columns=info_columns, # column(s) containing information about the column we want to impute
    output_column= impute_col[1], # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )
```

```python
imputer.fit(train_df=df_train, num_epochs=100)
```

```python
imputed = imputer.predict(data)
```

```python
imputed
```

```python
data["corpus"] = imputed["corpus_imputed"]
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.feature_extraction import DictVectorizer as DV
```

```python
vectorizer = DV(sparse=False)
str_columns = ["Sex", "corpus", "first_name", "second_name", "title"]
num_columns = [ "Age", "number_of_cabin","Pclass", "SibSp", "Parch", "Fare", "ticket"]
data_cat = vectorizer.fit_transform(data[str_columns].T.to_dict().values())
sc = StandardScaler()
sc_data_num = sc.fit_transform(data[num_columns].values)
data_num = np.column_stack((sc_data_num, data_cat))
```

```python
import sys
from impyute.imputation.cs import fast_knn
sys.setrecursionlimit(100000) #Increase the recursion limit of the OS

# start the KNN training
imputed_training=fast_knn(data_num, k=10)
```

```python
imputed_training.shape
```

```python
sc_data_num.shape
```

```python
ans = sc.inverse_transform(imputed_training[:,:7])
```

```python
ans[:, 1]
```

```python
data["Age"] = ans[:, 0]
```

```python
data["number_of_cabin"] = ans[:, 1]
```

```python
data
```

```python
for col in data.columns:
    data_first[col] = data[col]
```
